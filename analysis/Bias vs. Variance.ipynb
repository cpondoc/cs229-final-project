## Import Libraries
For our purposes, we only need `matplotlib`.

```python
import matplotlib.pyplot as plt
```

## Registering Manual Data
We manually input the accuracies from convergence from training as well as the test accuracies.

```python
network = 'Transfer Learning'
ai_model = 'Stable Diffusion'
props = [0.5, 0.6, 0.7, 0.8, 0.9]
train_accs = [98.97124481201172, 99.09109497070312, 99.33605194091797, 99.233154296875, 99.02230834960938]
test_accs = [96.58808933002481, 97.39454094292805, 97.02233250620347, 97.02233250620347, 97.39454094292805]
```

## Generating Graphs
We plot the train and test accuracies against one another to get a feel for bias and variance.

```python
fig = plt.figure()
fig, ax = plt.subplots()
ax.plot(props, train_accs, label="train accuracy")
ax.plot(props, test_accs, label="test accuracy")
ax.set_xlabel('Proportion of Training Data')
ax.set_ylabel('Percentage Accuracy')
ax.set_title(network + " vs. Size of Training Data for " + ai_model)
ax.legend()
```
